Simple linear regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
df = pd.read_csv("Salary_Data.csv")
df.head()
```

```python
#the iloc func is to select rows from the dataframe
x = df.iloc[:,:1].values #locate YearsExperience
y = df.iloc[:,-1].values #locate salary column

plt.scatter(x,y,color = 'm')
plt.xlabel('Independent Variable')
plt.ylabel('Dependent Variable')
plt.show()
```

```python
salary_model = LinearRegression()
salary_model.fit(x,y)
```

```python
y_pred = salary_model.predict(x)

plt.scatter(x,y,color = 'b')
plt.plot(x,y_pred, color = 'g')
plt.ylabel('Salary')
plt.xlabel('YearsExperience')
plt.show()

#prediction in green-line of best fit, observed values in blue.
```

```python
#introducing the calculation an employee who has been at the company longer than the previous employer
older_x = [[12]]

#simple linear regression takes in only 1-d arrays
x_pred = np.append(x, older_x).reshape(-1,1)
y_pred = salary_model.predict(x_pred)

plt.scatter(x,y,color = 'r')
plt.plot(x_pred,y_pred, color = 'g')
plt.ylabel('Salary')
plt.xlabel('YearsExperience')
plt.show()

print('Employee should make:', salary_model.predict(older_x))

#print(y_pred) - for all the predicted data
```
